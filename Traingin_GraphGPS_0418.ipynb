```python
def seed_everything(seed):
    # 固定亂數種子，讓每次實驗結果更可重現。
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)


def resolve_resplan_pkl_path(config):
    # 根據 cfg["data_source"] 決定這次資料到底從哪裡讀。
    data_source = config["data_source"].lower()
    if data_source == "kaggle":
        root = config["kaggle_dataset_root"]
        pkl_path = os.path.join(root, "ResPlan.pkl")
    elif data_source == "local":
        root = config["local_dataset_root"]
        pkl_path = os.path.join(root, "ResPlan.pkl")
    elif data_source == "custom":
        pkl_path = config["resplan_pkl_path"]
        root = os.path.dirname(pkl_path) if pkl_path else ""
    else:
        raise ValueError(f"Unsupported data_source: {config['data_source']}")

    if not pkl_path:
        raise ValueError("resplan_pkl_path is empty. Set cfg['resplan_pkl_path'] when using custom data_source.")
    if not os.path.exists(pkl_path):
        raise FileNotFoundError(
            f"Resolved ResPlan.pkl not found at: {pkl_path}. "
            f"Check cfg['data_source'] and the corresponding dataset root/path settings."
        )
    return {
        "data_source": data_source,
        "dataset_root": root,
        "resplan_pkl_path": pkl_path,
    }


def load_plans_from_source(config):
    # 真正執行資料載入的地方。
    resolved = resolve_resplan_pkl_path(config)
    with open(resolved["resplan_pkl_path"], "rb") as f:
        runtime_plans = pickle.load(f)
    return resolved, runtime_plans


seed_everything(cfg["seed"])
resolved_data, plans_runtime = load_plans_from_source(cfg)
plans = plans_runtime

# 下面這些路徑會被 training / inference / evaluation 共用。
CHECKPOINT_DIR = cfg["checkpoint_dir"]
OUTPUT_DIR = cfg["output_dir"]
RUN_NAME = cfg["run_name"]
os.makedirs(CHECKPOINT_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)

print("Data source:", resolved_data["data_source"])
print("Dataset root:", resolved_data["dataset_root"])
print("ResPlan.pkl:", resolved_data["resplan_pkl_path"])
print("Loaded plans:", len(plans))
print("Checkpoint dir:", CHECKPOINT_DIR)
print("Output dir:", OUTPUT_DIR)
```

## 5) Data Conversion Helpers

```python
import re
import copy
import numpy as np

TYPE2ID = {
    "living": 1, "kitchen": 2, "bedroom": 3, "bathroom": 4,
    "balcony": 5, "front_door": 15, "door": 17, "interior_door": 17,
    "storage": 10, "unknown": 16,
}
NEED_DOOR_TYPES = {"adjacency", "via_door"}

def name_to_type(n):
    m = re.match(r"^(.+)_\d+$", str(n))
    return m.group(1) if m else "unknown"

def extract_with_door_nodes(G_raw):
    # ---- 0. deep copy + 去重 ----
    G = copy.deepcopy(G_raw)
    nodes_list = list(G.nodes(data=True))
    to_remove = []
    for i in range(len(nodes_list)):
        if nodes_list[i][0] in to_remove:
            continue
        gi = nodes_list[i][1].get("geometry")
        if gi is None:
            continue
        for j in range(i + 1, len(nodes_list)):
            if nodes_list[j][0] in to_remove:
                continue
            gj = nodes_list[j][1].get("geometry")
            if gj is None:
                continue
            if gi.equals(gj):
                to_remove.append(nodes_list[j][0])
    for node in to_remove:
        G.remove_node(node)

    # ---- 1. 只保留有邊的 node ----
    connected = set()
    for u, v in G.edges():
        connected.add(u)
        connected.add(v)
    original_nodes = [n for n in G.nodes() if n in connected]

    # ---- 2. 插入 interior_door node ----
    final_nodes = list(original_nodes)
    final_edges = []
    door_count = 0
    door_to_pair = {}

    for u, v, d in G.edges(data=True):
        if u not in connected or v not in connected:
            continue
        etype = d.get("type", "unknown")
        if etype in NEED_DOOR_TYPES:
            door_name = f"interior_door_{door_count}"
            door_count += 1
            final_nodes.append(door_name)
            final_edges.append((u, door_name))
            final_edges.append((door_name, v))
            door_to_pair[door_name] = (u, v)
        else:
            final_edges.append((u, v))

    # ---- 3. index mapping + room_type ----
    n2i = {n: i for i, n in enumerate(final_nodes)}
    room_type = [TYPE2ID.get(name_to_type(n), 16) for n in final_nodes]

    # ---- 4. pairwise triples ----
    pos = set()
    for u, v in final_edges:
        i, j = n2i[u], n2i[v]
        if i > j:
            i, j = j, i
        pos.add((i, j))

    N = len(final_nodes)
    triples = []
    for i in range(N):
        for j in range(i + 1, N):
            triples.append([i, 1 if (i, j) in pos else -1, j])

    return {
        "node_names": final_nodes,
        "room_type": room_type,
        "triples": triples,
        # "n_original": len(original_nodes),
        "n_doors_added": door_count,
        "n_dedup_removed": len(to_remove),
        "door_to_pair": door_to_pair,
    }

result = extract_with_door_nodes(plan["graph"])


print("去重移除:", result["n_dedup_removed"], "個 node")
print("node_names:", result["node_names"])
print("room_type:", result["room_type"])
print("N total:", len(result["node_names"]))
print("N doors added:", result["n_doors_added"])
print("door_to_pair:", result["door_to_pair"])
print("triples", result["triples"])

```

```python
import numpy as np
import cv2
from shapely import affinity
from resplan_utils import normalize_keys, get_geometries, geometry_to_mask

def compute_plan_transform(plan, canvas_size=256, padding=4):
    categories = ("living", "kitchen", "bedroom", "bathroom", "balcony", "front_door", "door", "storage")
    plan = normalize_keys(plan)
    all_bounds = []
    for cat in categories:
        for g in get_geometries(plan.get(cat)):
            if not g.is_empty:
                all_bounds.append(g.bounds)
    if not all_bounds:
        return 0, 0, 1.0
    arr = np.array(all_bounds)
    minx, miny = arr[:, 0].min(), arr[:, 1].min()
    maxx, maxy = arr[:, 2].max(), arr[:, 3].max()
    range_x = maxx - minx
    range_y = maxy - miny
    max_range = max(range_x, range_y)
    if max_range < 1e-6:
        return -minx, -miny, 1.0
    usable = canvas_size - 2 * padding
    scale = usable / max_range
    shift_x = -minx * scale + padding + (usable - range_x * scale) / 2
    shift_y = -miny * scale + padding + (usable - range_y * scale) / 2
    return shift_x, shift_y, scale

def transform_geom(geom, shift_x, shift_y, scale):
    g = affinity.scale(geom, xfact=scale, yfact=scale, origin=(0, 0))
    g = affinity.translate(g, xoff=shift_x, yoff=shift_y)
    return g

def build_rms_masks_aligned(
    plan,
    G,
    node_names,
    room_type,
    door_to_pair,
    out_size=64,
    canvas_size=256,
    door_buffer=3.0,
):
    plan = normalize_keys(plan)

    # ---- 0. 計算座標轉換參數 ----
    shift_x, shift_y, scale = compute_plan_transform(plan, canvas_size=canvas_size)

    # ---- 1. 建立 room name → geometry 的 lookup（已轉換座標） ----
    geom_lookup = {}
    for cat in ("living", "kitchen", "bedroom", "bathroom", "balcony", "front_door", "storage"):
        geoms = get_geometries(plan.get(cat))
        for i, g in enumerate(geoms):
            geom_lookup[f"{cat}_{i}"] = transform_geom(g, shift_x, shift_y, scale)

    for n, d in G.nodes(data=True):
        g = d.get("geometry", None)
        if g is not None and n not in geom_lookup:
            geom_lookup[n] = transform_geom(g, shift_x, shift_y, scale)

    # ---- 2. 取出所有 door polygon（已轉換座標） ----
    door_polys_raw = get_geometries(plan.get("door"))
    door_polys = [transform_geom(dp, shift_x, shift_y, scale) for dp in door_polys_raw]

    # ---- 3. 配對 interior_door → door polygon ----
    door_geom_map = {}
    used_door_polys = set()

    for door_name, (room_a, room_b) in door_to_pair.items():
        geom_a = geom_lookup.get(room_a)
        geom_b = geom_lookup.get(room_b)
        if geom_a is None or geom_b is None:
            continue

        best_poly = None
        best_dist = float("inf")

        for idx, dp in enumerate(door_polys):
            if idx in used_door_polys:
                continue
            touches_a = dp.buffer(door_buffer).intersects(geom_a)
            touches_b = dp.buffer(door_buffer).intersects(geom_b)
            if touches_a and touches_b:
                from shapely.geometry import Point
                mid_x = (geom_a.centroid.x + geom_b.centroid.x) / 2
                mid_y = (geom_a.centroid.y + geom_b.centroid.y) / 2
                dist = dp.centroid.distance(Point(mid_x, mid_y))
                if dist < best_dist:
                    best_dist = dist
                    best_poly = idx

        if best_poly is not None:
            door_geom_map[door_name] = door_polys[best_poly]
            used_door_polys.add(best_poly)

    # ---- 4. 對每個 node 建 mask ----
    masks = []
    for name in node_names:
        geom = door_geom_map.get(name, None)
        if geom is None:
            geom = geom_lookup.get(name, None)

        if geom is None or geom.is_empty:
            masks.append(np.zeros((out_size, out_size), dtype=np.float32))
        else:
            m256 = geometry_to_mask(geom, shape=(canvas_size, canvas_size))
            m64 = cv2.resize(m256, (out_size, out_size), interpolation=cv2.INTER_NEAREST)
            masks.append(np.flipud((m64 > 0).astype(np.float32)))


    return np.stack(masks, axis=0) if masks else np.zeros((0, out_size, out_size), dtype=np.float32)
```

## 6) Dataset, Collate, And Fixed-Node Conditioning

```python
from torch.utils.data import Dataset, DataLoader


def one_hot_embedding(labels, num_classes=19):
    y = torch.eye(num_classes)
    return y[labels]


def plan_to_sample(plan, out_size=64):
    # 把單一 plan 轉成訓練真正會吃的三個東西：
    # 1. masks_tensor: 每個 node 的 mask
    # 2. nodes_tensor: 每個 node 的 one-hot room type
    # 3. edges_tensor: graph triples
    G_raw = plan["graph"]
    result = extract_with_door_nodes(G_raw)
    rms_masks = build_rms_masks_aligned(
        plan,
        G_raw,
        result["node_names"],
        result["room_type"],
        result["door_to_pair"],
        out_size=out_size,
    )

    room_type = np.array(result["room_type"])
    triples = np.array(result["triples"], dtype=np.int64)
    masks_tensor = torch.FloatTensor(rms_masks)
    masks_tensor = (masks_tensor - 0.5) / 0.5
    nodes_tensor = one_hot_embedding(torch.LongTensor(room_type))[:, 1:]
    edges_tensor = torch.LongTensor(triples)
    return masks_tensor, nodes_tensor, edges_tensor


class ResPlanDataset(Dataset):
    # 這個 Dataset 只是把 plan list 包成 PyTorch Dataset，
    # 真正的資料轉換工作在 plan_to_sample()。
    def __init__(self, plans, out_size=64):
        self.plans = plans
        self.out_size = out_size

    def __len__(self):
        return len(self.plans)

    def __getitem__(self, idx):
        return plan_to_sample(self.plans[idx], out_size=self.out_size)


def resplan_collate_fn(samples):
    # 把不同 plan 的 node / edge 拼成一個 batch，並維護 node_offset。
    all_mks, all_nds, all_eds = [], [], []
    all_nd_to_sample, all_ed_to_sample = [], []
    node_offset = 0

    for i, (mks, nds, eds) in enumerate(samples):
        N = nds.size(0)
        E = eds.size(0)
        all_mks.append(mks)
        all_nds.append(nds)

        eds = eds.clone()
        if E > 0:
            eds[:, 0] += node_offset
            eds[:, 2] += node_offset
            all_eds.append(eds)

        all_nd_to_sample.append(torch.LongTensor(N).fill_(i))
        all_ed_to_sample.append(torch.LongTensor(E).fill_(i))
        node_offset += N

    all_mks = torch.cat(all_mks, 0)
    all_nds = torch.cat(all_nds, 0)
    all_eds = torch.cat(all_eds, 0) if all_eds else torch.zeros(0, 3, dtype=torch.long)
    all_nd_to_sample = torch.cat(all_nd_to_sample)
    all_ed_to_sample = torch.cat(all_ed_to_sample)
    return all_mks, all_nds, all_eds, all_nd_to_sample, all_ed_to_sample
```

```python
import torch
import numpy as np
import random

ROOM_CLASS = {
    "living_room": 1, "kitchen": 2, "bedroom": 3, "bathroom": 4,
    "balcony": 5, "entrance": 6, "dining room": 7, "study room": 8,
    "storage": 10, "front door": 15, "unknown": 16, "interior_door": 17
}

def selectNodesTypes(nd_to_sample, batch_size, nds, keep_prob=None):
    keep_prob = cfg.get("fixed_keep_prob", 0.5) if keep_prob is None else keep_prob
    if not isinstance(nd_to_sample, torch.Tensor):
        nd_to_sample = torch.as_tensor(nd_to_sample, dtype=torch.long)
    else:
        nd_to_sample = nd_to_sample.detach().cpu().long()
    if not isinstance(nds, torch.Tensor):
        nds = torch.as_tensor(nds)
    else:
        nds = nds.detach().cpu()

    all_types = torch.tensor([ROOM_CLASS[k] - 1 for k in ROOM_CLASS], dtype=torch.long)
    node_types = nds.argmax(dim=1).long()
    fixed_nodes = []
    for k in range(batch_size):
        sample_nodes = torch.nonzero(nd_to_sample == k, as_tuple=False).flatten()
        if sample_nodes.numel() == 0:
            continue
        chosen_types = all_types[torch.rand(all_types.numel()) < keep_prob]
        if chosen_types.numel() == 0:
            continue
        fixed_mask = torch.isin(node_types[sample_nodes], chosen_types)
        fixed_nodes.append(sample_nodes[fixed_mask])

    if fixed_nodes:
        fixed_nodes = torch.cat(fixed_nodes).long()
    else:
        fixed_nodes = torch.empty(0, dtype=torch.long)
    bin_fixed_nodes = torch.zeros((nd_to_sample.shape[0], 1), dtype=torch.float32)
    if fixed_nodes.numel() > 0:
        bin_fixed_nodes[fixed_nodes] = 1.0
    return fixed_nodes, bin_fixed_nodes

def fix_nodes(prev_mks, ind_fixed_nodes):
    if isinstance(prev_mks, torch.Tensor):
        given_masks = prev_mks.detach().clone()
    else:
        given_masks = torch.tensor(prev_mks, dtype=torch.float32)
    ind_not_fixed_nodes = torch.tensor(
        [k for k in range(given_masks.shape[0]) if k not in ind_fixed_nodes],
        dtype=torch.long
    )
    given_masks[ind_not_fixed_nodes] = -1.0
    given_masks = given_masks.unsqueeze(1)
    inds_masks = torch.zeros_like(given_masks)
    inds_masks[ind_not_fixed_nodes] = 0.0
    if isinstance(ind_fixed_nodes, torch.Tensor):
        inds_masks[ind_fixed_nodes.long()] = 1.0
    else:
        inds_masks[torch.tensor(ind_fixed_nodes, dtype=torch.long)] = 1.0
    given_masks = torch.cat([given_masks, inds_masks], 1)
    return given_masks

def _init_input(graph, prev_state=None, mask_size=64):
    given_nds, given_eds = graph
    given_nds = given_nds.float()
    if isinstance(given_eds, torch.Tensor):
        given_eds = given_eds.detach().clone().long()
    else:
        given_eds = torch.tensor(given_eds, dtype=torch.long)
    z = torch.randn(len(given_nds), 128, dtype=torch.float32)
    fixed_nodes = prev_state["fixed_nodes"]
    if prev_state["masks"] is None:
        prev_mks = torch.zeros((given_nds.shape[0], mask_size, mask_size), dtype=torch.float32) - 1.0
    else:
        prev_mks = prev_state["masks"]
        if isinstance(prev_mks, torch.Tensor):
            prev_mks = prev_mks.detach().clone()
        else:
            prev_mks = torch.tensor(prev_mks, dtype=torch.float32)
    if isinstance(fixed_nodes, torch.Tensor):
        fixed_nodes_t = fixed_nodes.detach().clone()
    else:
        fixed_nodes_t = torch.tensor(fixed_nodes, dtype=torch.long)
    given_masks_in = fix_nodes(prev_mks, fixed_nodes_t)
    if torch.cuda.is_available():
        z = z.cuda()
        given_masks_in = given_masks_in.cuda()
    return z, given_masks_in, given_nds, given_eds
```

```python
import inspect
import json
import math
import shutil
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm import tqdm


def _read_secret_if_available(secret_name):
    # Kaggle Notebook 上可以從 Add-ons -> Secrets 讀取 API key。
    if UserSecretsClient is None:
        return None
    try:
        return UserSecretsClient().get_secret(secret_name)
    except Exception:
        return None


def resolve_wandb_api_key(config):
    if os.environ.get("WANDB_API_KEY"):
        return os.environ["WANDB_API_KEY"]
    return _read_secret_if_available("WANDB_API_KEY")


def make_wandb_stage_config(config, stage_name, run_suffix):
    stage_config = dict(config)
    stage_config["training_stage"] = stage_name
    # stage_config["wandb_run_name"] = f"{config.get('run_name', 'experiment')}_{run_suffix}"
    return stage_config


def init_wandb_if_needed(config, generator=None, discriminator=None):
    if not config.get("use_wandb", False):
        print("W&B disabled.")
        return None
    if wandb is None:
        print("wandb is not installed. Skipping W&B logging.")
        return None

    api_key = resolve_wandb_api_key(config)
    if not api_key:
        print("WANDB_API_KEY not found. Skipping W&B logging.")
        return None

    try:
        wandb.login(key=api_key)
        run_name = config.get("wandb_run_name") or config.get("run_name", "experiment")
        init_kwargs = {
            "project": config.get("wandb_project", "resplan-graphgps"),
            "name": run_name,
            "config": config,
            "dir": os.path.join(config["output_dir"], config["run_name"]),
            "mode": config.get("wandb_mode", "online"),
            
            # "id": "u61aiqbm",
            # "resume" : "must"
        }
        if config.get("wandb_entity"):
            init_kwargs["entity"] = config["wandb_entity"]
        run = wandb.init(**init_kwargs)
        if config.get("wandb_watch_model", False):
            if generator is not None:
                wandb.watch(generator, log="all", log_freq=100)
            if discriminator is not None:
                wandb.watch(discriminator, log="all", log_freq=100)
        print("W&B run started:", run.name)
        return run
    except Exception as exc:
        print(f"W&B init failed: {exc}")
        return None


def checkpoint_paths(checkpoint_dir, run_name):
    run_dir = Path(checkpoint_dir) / run_name
    return {
        "root": str(run_dir),
        "last": str(run_dir / "last.pth"),
        "best": str(run_dir / "best.pth"),
        "config": str(run_dir / "config.yaml"),
        "model_source": str(run_dir / "graphgps_models.py"),
    }


def epoch_checkpoint_path(checkpoint_dir, run_name, epoch):
    return str(Path(checkpoint_paths(checkpoint_dir, run_name)["root"]) / f"epoch_{int(epoch):03d}.pth")


def _legacy_checkpoint_paths(checkpoint_dir, run_name):
    return {
        "last": os.path.join(checkpoint_dir, f"{run_name}_last.pth"),
        "best": os.path.join(checkpoint_dir, f"{run_name}_best.pth"),
    }


def _extract_model_kwargs(module):
    fields = ("channels", "noise_dim", "cond_dim")
    kwargs = {field: getattr(module, field) for field in fields if hasattr(module, field)}
    return kwargs or None


def _load_state_dict_with_model_hint(module, state_dict, checkpoint_model_name=None, strict=True):
    try:
        return module.load_state_dict(state_dict, strict=strict)
    except RuntimeError as exc:
        expected_model = MODEL_NAME
        mismatch_lines = [str(exc)]
        if checkpoint_model_name and checkpoint_model_name != expected_model:
            mismatch_lines.append(
                f"Checkpoint model_name={checkpoint_model_name!r}, but current notebook expects {expected_model!r}."
            )
        else:
            mismatch_lines.append(
                f"Current notebook expects model_name={expected_model!r}. This usually means the checkpoint came from a different architecture version."
            )
        raise RuntimeError("\n".join(mismatch_lines)) from exc


def _yaml_scalar(value):
    if value is None:
        return "null"
    if isinstance(value, bool):
        return "true" if value else "false"
    if isinstance(value, (int, float)):
        return repr(value)
    return json.dumps(str(value), ensure_ascii=False)


def _to_yaml_lines(value, indent=0):
    prefix = "  " * indent
    if isinstance(value, dict):
        lines = []
        for key, item in value.items():
            if isinstance(item, (dict, list)):
                lines.append(f"{prefix}{key}:")
                lines.extend(_to_yaml_lines(item, indent + 1))
            else:
                lines.append(f"{prefix}{key}: {_yaml_scalar(item)}")
        return lines or [f"{prefix}{{}}"]
    if isinstance(value, list):
        lines = []
        for item in value:
            if isinstance(item, (dict, list)):
                lines.append(f"{prefix}-")
                lines.extend(_to_yaml_lines(item, indent + 1))
            else:
                lines.append(f"{prefix}- {_yaml_scalar(item)}")
        return lines or [f"{prefix}[]"]
    return [f"{prefix}{_yaml_scalar(value)}"]


def _serialize_config_to_yaml(config):
    return "\n".join(_to_yaml_lines(config)) + "\n"


def _ensure_checkpoint_metadata(config, run_name, generator=None):
    paths = checkpoint_paths(config["checkpoint_dir"], run_name)
    root = Path(paths["root"])
    root.mkdir(parents=True, exist_ok=True)

    config_path = Path(paths["config"])
    config_path.write_text(_serialize_config_to_yaml(config), encoding="utf-8")

    model_source_path = None
    if generator is not None:
        model_source_path = _resolve_model_source_path(generator.__class__)
    if model_source_path and os.path.exists(model_source_path):
        shutil.copyfile(model_source_path, paths["model_source"])

    return paths


def _resolve_model_source_path(module_or_class):
    source_path = inspect.getsourcefile(module_or_class)
    if source_path is None:
        return None
    return str(Path(source_path).resolve())


def upload_wandb_training_artifact(wandb_run, config, run_name, generator, checkpoint_path, epoch, metrics=None):
    if wandb_run is None or wandb is None:
        return None
    if not config.get("upload_best_artifact_to_wandb", True):
        return None
    checkpoint_path = str(Path(checkpoint_path).resolve())
    if not os.path.exists(checkpoint_path):
        print(f"Best checkpoint not found for W&B artifact upload: {checkpoint_path}")
        return None

    paths = _ensure_checkpoint_metadata(config, run_name, generator=generator)
    config_path = paths["config"]
    model_source_path = paths["model_source"] if os.path.exists(paths["model_source"]) else None

    metadata = {
        "model_name": MODEL_NAME,
        "run_name": run_name,
        "epoch": int(epoch),
    }
    if metrics is not None:
        for key, value in metrics.items():
            if isinstance(value, (int, float, str, bool)):
                metadata[key] = value

    artifact = wandb.Artifact(
        name=f"{run_name}-best-model",
        type="model",
        metadata=metadata,
    )
    artifact.add_file(checkpoint_path, name=os.path.basename(checkpoint_path))
    artifact.add_file(config_path, name="config.yaml")
    if model_source_path and os.path.exists(model_source_path):
        artifact.add_file(model_source_path, name=Path(model_source_path).name)

    aliases = ["best", f"epoch-{int(epoch):03d}"]
    logged_artifact = wandb_run.log_artifact(artifact, aliases=aliases)
    if hasattr(wandb_run, "summary"):
        wandb_run.summary["best_artifact_name"] = artifact.name
        wandb_run.summary["best_artifact_checkpoint"] = checkpoint_path
        if model_source_path:
            wandb_run.summary["best_artifact_model_source"] = model_source_path
        wandb_run.summary["best_artifact_config"] = config_path
    return logged_artifact


def save_checkpoint(generator, discriminator, optimizer_G, optimizer_D, epoch, global_step, config, metrics, checkpoint_dir, run_name, is_best=False):
    paths = _ensure_checkpoint_metadata(config, run_name, generator=generator)
    checkpoint = {
        "epoch": epoch,
        "global_step": global_step,
        "model_name": MODEL_NAME,
        "model_kwargs": _extract_model_kwargs(generator),
        "generator_state_dict": generator.state_dict(),
        "discriminator_state_dict": discriminator.state_dict(),
        "optimizer_G_state_dict": optimizer_G.state_dict(),
        "optimizer_D_state_dict": optimizer_D.state_dict(),
        "config": config,
        "metrics": metrics,
    }
    torch.save(checkpoint, paths["last"])
    if config.get("save_every_n_epochs") and epoch % config["save_every_n_epochs"] == 0:
        torch.save(checkpoint, epoch_checkpoint_path(checkpoint_dir, run_name, epoch))
    if is_best:
        torch.save(checkpoint, paths["best"])
    return paths


def save_pretrain_checkpoint(generator, optimizer_G, epoch, global_step, config, metrics, checkpoint_path):
    checkpoint_path = Path(checkpoint_path)
    checkpoint_path.parent.mkdir(parents=True, exist_ok=True)
    epoch_checkpoint_path = checkpoint_path.with_name(
        f"{checkpoint_path.stem}_epoch_{epoch:03d}{checkpoint_path.suffix}"
    )
    checkpoint = {
        "epoch": epoch,
        "global_step": global_step,
        "model_name": MODEL_NAME,
        "model_kwargs": _extract_model_kwargs(generator),
        "stage": "generator_pretrain",
        "latest_checkpoint_path": str(checkpoint_path),
        "epoch_checkpoint_path": str(epoch_checkpoint_path),
        "generator_state_dict": generator.state_dict(),
        "optimizer_G_state_dict": optimizer_G.state_dict(),
        "config": config,
        "metrics": metrics,
    }
    torch.save(checkpoint, checkpoint_path)
    torch.save(checkpoint, epoch_checkpoint_path)
    return str(checkpoint_path)


def load_pretrained_generator_checkpoint(checkpoint_path, generator, map_location="cpu"):
    checkpoint = torch.load(checkpoint_path, map_location=map_location)
    state_dict = checkpoint.get("generator_state_dict", checkpoint)
    checkpoint_model_name = checkpoint.get("model_name") if isinstance(checkpoint, dict) else None
    _load_state_dict_with_model_hint(
        generator,
        state_dict,
        checkpoint_model_name=checkpoint_model_name,
        strict=True,
    )
    return checkpoint


def load_training_checkpoint(checkpoint_path, generator, discriminator=None, optimizer_G=None, optimizer_D=None, map_location="cpu"):
    checkpoint = torch.load(checkpoint_path, map_location=map_location)
    checkpoint_model_name = checkpoint.get("model_name") if isinstance(checkpoint, dict) else None
    if "generator_state_dict" in checkpoint:
        _load_state_dict_with_model_hint(
            generator,
            checkpoint["generator_state_dict"],
            checkpoint_model_name=checkpoint_model_name,
            strict=True,
        )
        if discriminator is not None and "discriminator_state_dict" in checkpoint:
            _load_state_dict_with_model_hint(
                discriminator,
                checkpoint["discriminator_state_dict"],
                checkpoint_model_name=checkpoint_model_name,
                strict=True,
            )
        if optimizer_G is not None and "optimizer_G_state_dict" in checkpoint:
            optimizer_G.load_state_dict(checkpoint["optimizer_G_state_dict"])
        if optimizer_D is not None and "optimizer_D_state_dict" in checkpoint:
            optimizer_D.load_state_dict(checkpoint["optimizer_D_state_dict"])
        start_epoch = checkpoint.get("epoch", 0) + 1
        global_step = checkpoint.get("global_step", 0)
        metrics = checkpoint.get("metrics", {})
    else:
        _load_state_dict_with_model_hint(
            generator,
            checkpoint,
            checkpoint_model_name=checkpoint_model_name,
            strict=True,
        )
        start_epoch = 0
        global_step = 0
        metrics = {}
    return start_epoch, global_step, metrics


def resolve_checkpoint_path(selection, checkpoint_dir, run_name):
    paths = checkpoint_paths(checkpoint_dir, run_name)
    legacy_paths = _legacy_checkpoint_paths(checkpoint_dir, run_name)
    if selection in (None, "best"):
        for candidate in (paths["best"], paths["last"], legacy_paths["best"], legacy_paths["last"]):
            if os.path.exists(candidate):
                return candidate
        return paths["best"]
    if selection == "last":
        return paths["last"] if os.path.exists(paths["last"]) else legacy_paths["last"]
    if os.path.exists(str(selection)):
        return str(selection)
    raise FileNotFoundError(f"Could not resolve checkpoint from selection: {selection}")


def resolve_generator_checkpoint_path(selection, checkpoint_dir, run_name):
    return resolve_checkpoint_path(selection, checkpoint_dir, run_name)


def load_generator_from_checkpoint(generator, checkpoint_path, device, strict=True):
    checkpoint = torch.load(checkpoint_path, map_location=device)
    state_dict = checkpoint.get("generator_state_dict", checkpoint)
    checkpoint_model_name = checkpoint.get("model_name") if isinstance(checkpoint, dict) else None
    _load_state_dict_with_model_hint(
        generator,
        state_dict,
        checkpoint_model_name=checkpoint_model_name,
        strict=strict,
    )
    generator.to(device)
    generator.eval()
    return checkpoint


def log_metrics(metrics, step=None):
    if wandb_run is not None:
        payload = dict(metrics)
        if step is not None:
            payload["trainer/global_step"] = step
        wandb_run.log(payload)


def save_preview_grid(real_batch, fake_batch, out_path, fixed_nodes=None, max_items=6):
    fixed_nodes = set() if fixed_nodes is None else set(map(int, fixed_nodes))
    real_np = real_batch.detach().cpu().numpy()
    fake_np = fake_batch.detach().cpu().numpy()
    n_items = min(max_items, real_np.shape[0])
    fig, axes = plt.subplots(2, n_items, figsize=(2.5 * n_items, 5))
    if n_items == 1:
        axes = np.array(axes).reshape(2, 1)
    for idx in range(n_items):
        fixed_suffix = " fixed" if idx in fixed_nodes else ""
        axes[0, idx].imshow(real_np[idx], cmap="gray", vmin=-1, vmax=1)
        axes[0, idx].set_title(f"Real {idx}{fixed_suffix}")
        axes[0, idx].axis("off")
        axes[1, idx].imshow(fake_np[idx], cmap="gray", vmin=-1, vmax=1)
        axes[1, idx].set_title(f"Gen {idx}{fixed_suffix}")
        axes[1, idx].axis("off")
    axes[0, 0].set_ylabel("Real")
    axes[1, 0].set_ylabel("Gen")
    plt.tight_layout()
    fig.savefig(out_path, bbox_inches="tight")
    plt.close(fig)
    return out_path


wandb_run = None
PREVIEW_DIR = os.path.join(OUTPUT_DIR, RUN_NAME, "previews")
os.makedirs(PREVIEW_DIR, exist_ok=True)
print("Preview dir:", PREVIEW_DIR)
```

## 10) Losses And Optional Pretrain

```python
# ===== Optional Generator pretrain =====
distance_loss = torch.nn.L1Loss()

def weighted_l1(pred, target, pos_weight=15.0):
    weights = torch.ones_like(target)
    weights = torch.where(target > 0, weights * pos_weight, weights)
    return (weights * (pred - target).abs()).mean()

def _mask01(x):
    return ((x + 1.0) / 2.0).clamp(0.0, 1.0)

def dice_loss_tanh(pred, target, eps=1e-6):
    pred01 = _mask01(pred)
    target01 = _mask01(target)
    intersection = (pred01 * target01).sum(dim=(-2, -1))
    denom = pred01.sum(dim=(-2, -1)) + target01.sum(dim=(-2, -1))
    dice = (2.0 * intersection + eps) / (denom + eps)
    return 1.0 - dice.mean()

def foreground_ratio_loss_tanh(pred, target):
    pred_ratio = _mask01(pred).mean(dim=(-2, -1))
    target_ratio = _mask01(target).mean(dim=(-2, -1))
    return (pred_ratio - target_ratio).abs().mean()

def mask_reconstruction_loss(pred, target, config):
    wl1 = weighted_l1(pred, target, pos_weight=config["pretrain_foreground_weight"])
    dice = dice_loss_tanh(pred, target)
    fg_ratio = foreground_ratio_loss_tanh(pred, target)
    total = wl1 + config["pretrain_dice_weight"] * dice + config["pretrain_fg_ratio_weight"] * fg_ratio
    return total, wl1, dice, fg_ratio

def hard_copy_fixed_masks(gen_mks, real_mks, fixed_nodes):
    if fixed_nodes is None or len(fixed_nodes) == 0:
        return gen_mks
    fixed_nodes = fixed_nodes.long()
    conditioned = gen_mks.clone()
    conditioned[fixed_nodes] = real_mks[fixed_nodes]
    return conditioned

def nonfixed_reconstruction_loss(pred, target, fixed_nodes, config):
    fixed_mask = torch.zeros(pred.shape[0], dtype=torch.bool, device=pred.device)
    if fixed_nodes is not None and len(fixed_nodes) > 0:
        fixed_mask[fixed_nodes.long()] = True
    nonfixed_nodes = torch.arange(pred.shape[0], device=pred.device)[~fixed_mask]
    if len(nonfixed_nodes) == 0:
        zero = torch.tensor(0.0, device=pred.device)
        return zero, zero, zero, zero, zero
    recon, wl1, dice, fg_ratio = mask_reconstruction_loss(pred[nonfixed_nodes], target[nonfixed_nodes], config)
    gen_nonfixed_fg = (pred[nonfixed_nodes] > 0).float().mean()
    real_nonfixed_fg = (target[nonfixed_nodes] > 0).float().mean()
    empty_penalty = torch.relu(real_nonfixed_fg * 0.5 - gen_nonfixed_fg)
    return recon, wl1, dice, fg_ratio, empty_penalty

def render_fixed_preview(model, batch, out_path, hard_copy_fixed=False):
    was_training = model.training
    model.eval()
    with torch.no_grad():
        mks, nds, eds, nd_to_sample, ed_to_sample = batch
        batch_size = int(nd_to_sample.max().item()) + 1
        real_mks = mks.to(device, non_blocking=True)
        given_nds = nds.to(device, non_blocking=True)
        given_eds = eds.to(device, non_blocking=True)
        graph = [given_nds, given_eds]
        ind_fixed_nodes, _ = selectNodesTypes(nd_to_sample, batch_size, nds)
        ind_fixed_nodes = ind_fixed_nodes.to(device, non_blocking=True)
        state = {"masks": real_mks, "fixed_nodes": ind_fixed_nodes}
        z, given_masks_in, given_nds, given_eds = _init_input(graph, state)
        gen_mks = model(
            z.to(device, non_blocking=True),
            given_m=given_masks_in.to(device, non_blocking=True),
            given_y=given_nds.to(device, non_blocking=True),
            given_w=given_eds.to(device, non_blocking=True),
        )
        if hard_copy_fixed:
            gen_mks = hard_copy_fixed_masks(gen_mks, real_mks, ind_fixed_nodes)
        save_preview_grid(real_mks, gen_mks, out_path, fixed_nodes=ind_fixed_nodes)
    if was_training:
        model.train()
    return out_path

fixed_preview_batch = next(iter(fp_loader))

if cfg.get("run_pretrain", False):
    pretrain_generator = Generator().to(device)
    pretrain_optimizer_G = torch.optim.Adam(
        pretrain_generator.parameters(),
        lr=cfg["pretrain_lr"],
        betas=(cfg["b1"], cfg["b2"]),
    )
    pretrain_wandb_config = make_wandb_stage_config(cfg, stage_name="generator_pretrain", run_suffix="g_pretrain")
    wandb_run = init_wandb_if_needed(pretrain_wandb_config, generator=pretrain_generator)
    pretrain_global_step = 0
    best_pretrain_loss = math.inf

    for pretrain_epoch in range(cfg["pretrain_epochs"]):
        pretrain_generator.train()
        epoch_pretrain_loss = 0.0
        epoch_l1_all = 0.0
        epoch_l1_fixed = 0.0
        epoch_l1_nonfixed = 0.0
        epoch_empty_penalty = 0.0
        epoch_dice_all = 0.0
        epoch_fg_ratio_all = 0.0
        n_pretrain_batches = 0
        preview_path = None

        pbar = tqdm(
            enumerate(fp_loader),
            total=len(fp_loader),
            desc=f"G pretrain {pretrain_epoch + 1}/{cfg['pretrain_epochs']}",
            unit="batch",
            ncols=120,
        )

        for batch_idx, batch in pbar:
            mks, nds, eds, nd_to_sample, ed_to_sample = batch
            batch_size = int(nd_to_sample.max().item()) + 1
            real_mks = mks.to(device, non_blocking=True)
            given_nds = nds.to(device, non_blocking=True)
            given_eds = eds.to(device, non_blocking=True)

            graph = [given_nds, given_eds]
            ind_fixed_nodes, _ = selectNodesTypes(nd_to_sample, batch_size, nds)
            ind_fixed_nodes = ind_fixed_nodes.to(device, non_blocking=True)
            nd_to_sample = nd_to_sample.to(device, non_blocking=True)
            state = {"masks": real_mks, "fixed_nodes": ind_fixed_nodes}
            z, given_masks_in, given_nds, given_eds = _init_input(graph, state)
            z = z.to(device, non_blocking=True)
            given_masks_in = given_masks_in.to(device, non_blocking=True)
            given_nds = given_nds.to(device, non_blocking=True)
            given_eds = given_eds.to(device, non_blocking=True)

            pretrain_optimizer_G.zero_grad()
            gen_mks = pretrain_generator(z, given_m=given_masks_in, given_y=given_nds, given_w=given_eds)
            recon_all, l1_all, dice_all, fg_ratio_all = mask_reconstruction_loss(gen_mks, real_mks, cfg)
            if len(ind_fixed_nodes) > 0:
                l1_fixed = weighted_l1(
                    gen_mks[ind_fixed_nodes],
                    given_masks_in[ind_fixed_nodes, 0],
                    pos_weight=cfg["pretrain_foreground_weight"],
                )
            else:
                l1_fixed = torch.tensor(0.0, device=device)

            fixed_mask = torch.zeros(gen_mks.shape[0], dtype=torch.bool, device=device)
            if len(ind_fixed_nodes) > 0:
                fixed_mask[ind_fixed_nodes.long()] = True
            nonfixed_nodes = torch.arange(gen_mks.shape[0], device=device)[~fixed_mask]
            if len(nonfixed_nodes) > 0:
                l1_nonfixed = weighted_l1(
                    gen_mks[nonfixed_nodes],
                    real_mks[nonfixed_nodes],
                    pos_weight=cfg["pretrain_foreground_weight"],
                )
                gen_nonfixed_fg = (gen_mks[nonfixed_nodes] > 0).float().mean()
                real_nonfixed_fg = (real_mks[nonfixed_nodes] > 0).float().mean()
                empty_penalty = torch.relu(real_nonfixed_fg * 0.5 - gen_nonfixed_fg)
            else:
                l1_nonfixed = torch.tensor(0.0, device=device)
                empty_penalty = torch.tensor(0.0, device=device)

            pretrain_loss = (
                cfg["pretrain_l1_weight_all"] * recon_all
                + cfg["pretrain_l1_weight_fixed"] * l1_fixed
                + cfg["pretrain_l1_weight_nonfixed"] * l1_nonfixed
                + cfg["pretrain_empty_penalty_weight"] * empty_penalty
            )
            pretrain_loss.backward()
            pretrain_optimizer_G.step()

            epoch_pretrain_loss += pretrain_loss.item()
            epoch_l1_all += l1_all.item()
            epoch_l1_fixed += l1_fixed.item()
            epoch_l1_nonfixed += l1_nonfixed.item()
            epoch_empty_penalty += empty_penalty.item()
            epoch_dice_all += dice_all.item()
            epoch_fg_ratio_all += fg_ratio_all.item()
            n_pretrain_batches += 1

            log_metrics(
                {
                    "pretrain/loss": pretrain_loss.item(),
                    "pretrain/l1_all": l1_all.item(),
                    "pretrain/l1_fixed": l1_fixed.item(),
                    "pretrain/l1_nonfixed": l1_nonfixed.item(),
                    "pretrain/empty_penalty": empty_penalty.item(),
                    "pretrain/dice_all": dice_all.item(),
                    "pretrain/fg_ratio_loss_all": fg_ratio_all.item(),
                    "pretrain/real_foreground_ratio": (real_mks > 0).float().mean().item(),
                    "pretrain/gen_foreground_ratio": (gen_mks > 0).float().mean().item(),
                    "pretrain/epoch": pretrain_epoch + 1,
                },
                step=pretrain_global_step,
            )

            if preview_path is None:
                preview_path = os.path.join(PREVIEW_DIR, f"pretrain_epoch_{pretrain_epoch + 1:03d}.png")
                render_fixed_preview(pretrain_generator, fixed_preview_batch, preview_path)
                if wandb_run is not None:
                    wandb_run.log({"pretrain/preview": wandb.Image(preview_path), "trainer/global_step": pretrain_global_step})

            pbar.set_postfix(loss=pretrain_loss.item(), l1_all=l1_all.item(), l1_fixed=l1_fixed.item(), l1_nonfixed=l1_nonfixed.item())
            pretrain_global_step += 1

        avg_pretrain_loss = epoch_pretrain_loss / max(n_pretrain_batches, 1)
        avg_l1_all = epoch_l1_all / max(n_pretrain_batches, 1)
        avg_l1_fixed = epoch_l1_fixed / max(n_pretrain_batches, 1)
        avg_l1_nonfixed = epoch_l1_nonfixed / max(n_pretrain_batches, 1)
        avg_empty_penalty = epoch_empty_penalty / max(n_pretrain_batches, 1)
        avg_dice_all = epoch_dice_all / max(n_pretrain_batches, 1)
        avg_fg_ratio_all = epoch_fg_ratio_all / max(n_pretrain_batches, 1)
        pretrain_metrics = {
            "epoch": pretrain_epoch + 1,
            "avg_pretrain_loss": avg_pretrain_loss,
            "avg_l1_all": avg_l1_all,
            "avg_l1_fixed": avg_l1_fixed,
            "avg_l1_nonfixed": avg_l1_nonfixed,
            "avg_empty_penalty": avg_empty_penalty,
            "avg_dice_all": avg_dice_all,
            "avg_fg_ratio_all": avg_fg_ratio_all,
            "lr_g": pretrain_optimizer_G.param_groups[0]["lr"],
        }
        log_metrics(
            {
                "pretrain_epoch/avg_loss": avg_pretrain_loss,
                "pretrain_epoch/avg_l1_all": avg_l1_all,
                "pretrain_epoch/avg_l1_fixed": avg_l1_fixed,
                "pretrain_epoch/avg_l1_nonfixed": avg_l1_nonfixed,
                "pretrain_epoch/avg_empty_penalty": avg_empty_penalty,
                "pretrain_epoch/avg_dice_all": avg_dice_all,
                "pretrain_epoch/avg_fg_ratio_all": avg_fg_ratio_all,
                "pretrain_epoch/lr_g": pretrain_optimizer_G.param_groups[0]["lr"],
                "pretrain_epoch/epoch": pretrain_epoch + 1,
            },
            step=pretrain_global_step,
        )
        best_pretrain_loss = min(best_pretrain_loss, avg_pretrain_loss)
        checkpoint_path = save_pretrain_checkpoint(
            pretrain_generator,
            pretrain_optimizer_G,
            pretrain_epoch + 1,
            pretrain_global_step,
            cfg,
            pretrain_metrics,
            cfg["pretrain_checkpoint"],
        )
        print(
            f"Pretrain epoch {pretrain_epoch + 1}: "
            f"avg_loss={avg_pretrain_loss:.4f}, avg_l1_all={avg_l1_all:.4f}, "
            f"avg_l1_fixed={avg_l1_fixed:.4f}"
        )
        print(
            "Pretrain epoch checkpoint:",
            Path(cfg["pretrain_checkpoint"]).with_name(
                f"{Path(cfg['pretrain_checkpoint']).stem}_epoch_{pretrain_epoch + 1:03d}{Path(cfg['pretrain_checkpoint']).suffix}"
            ),
        )

    print("Generator pretrain complete.")
    print("Pretrain checkpoint:", cfg["pretrain_checkpoint"])
    if wandb_run is not None:
        wandb_run.summary["best_pretrain_loss"] = best_pretrain_loss
        wandb_run.summary["pretrain_checkpoint"] = cfg["pretrain_checkpoint"]
        wandb_run.finish()
    wandb_run = None
```

```python
distance_loss = torch.nn.L1Loss()

if not cfg.get("run_main_training", True):
    print("Skipping main training for preset:", selected_preset)
else:
        # ===== 建立 WGAN-GP 模型與 optimizer =====
    # 這裡開始是真正的 WGAN-GP 訓練主流程。
    generator = Generator().to(device)
    discriminator = Discriminator().to(device)
    optimizer_G = torch.optim.Adam(generator.parameters(), lr=cfg["g_lr"], betas=(cfg["b1"], cfg["b2"]))
    optimizer_D = torch.optim.Adam(discriminator.parameters(), lr=cfg["d_lr"], betas=(cfg["b1"], cfg["b2"]))

    resume_path = cfg["resume_checkpoint"]
    if resume_path:
        print("resume_checkpoint is set; skipping pretrained-generator initialization and resuming full GAN state.")
    elif cfg.get("load_pretrained_generator", False):
        pretrained_generator_path = cfg.get("pretrained_generator_checkpoint")
        if pretrained_generator_path and os.path.exists(pretrained_generator_path):
            _ = load_pretrained_generator_checkpoint(pretrained_generator_path, generator, map_location=device)
            print("Loaded pretrained generator:", pretrained_generator_path)
        else:
            print(f"Pretrained generator checkpoint not found: {pretrained_generator_path}. Starting G from scratch.")

    # 模型建立完成後再 init wandb，這樣需要的 model watch 才能接上。
    gan_wandb_config = make_wandb_stage_config(cfg, stage_name="wgangp_finetune", run_suffix="wgangp")
    wandb_run = init_wandb_if_needed(gan_wandb_config, generator=generator, discriminator=discriminator)

    start_epoch = 0
    global_step = 0
    best_metric_name = cfg.get("best_metric", "avg_g_loss")
    best_g_loss = math.inf
    if resume_path:
        resume_path = resolve_checkpoint_path(resume_path, checkpoint_dir=CHECKPOINT_DIR, run_name=RUN_NAME)
        start_epoch, global_step, resume_metrics = load_training_checkpoint(
            resume_path,
            generator,
            discriminator,
            optimizer_G,
            optimizer_D,
            map_location=device,
        )
        best_g_loss = resume_metrics.get(best_metric_name, math.inf)
        if cfg.get("reset_optimizer_lr_on_resume", True):
            for group in optimizer_G.param_groups:
                group["lr"] = cfg["g_lr"]
            for group in optimizer_D.param_groups:
                group["lr"] = cfg["d_lr"]
            print("Resume LR reset:", optimizer_G.param_groups[0]["lr"], optimizer_D.param_groups[0]["lr"])
        print(f"Resumed from {resume_path} at epoch {start_epoch}")
        log_metrics({"system/resumed_from_checkpoint": 1, "system/start_epoch": start_epoch}, step=global_step)

    for epoch in range(start_epoch, cfg["n_epochs"]):
        generator.train()
        discriminator.train()
        epoch_d_loss = 0.0
        epoch_g_loss = 0.0
        epoch_l1_loss = 0.0
        epoch_gp_loss = 0.0
        epoch_g_adv_loss = 0.0
        epoch_g_fg_ratio_loss = 0.0
        epoch_g_nonfixed_recon_loss = 0.0
        epoch_g_empty_penalty_loss = 0.0
        epoch_recon_all_loss = 0.0
        n_batches = 0
        n_d_steps = 0
        n_g_steps = 0
        preview_path = None

        pbar = tqdm(
            enumerate(fp_loader),
            total=len(fp_loader),
            desc=f"Epoch {epoch + 1}/{cfg['n_epochs']}",
            unit="batch",
            ncols=120,
        )

        for batch_idx, batch in pbar:
            mks, nds, eds, nd_to_sample, ed_to_sample = batch
            batch_size = int(nd_to_sample.max().item()) + 1
            real_mks = mks.to(device, non_blocking=True)
            given_nds = nds.to(device, non_blocking=True)
            given_eds = eds.to(device, non_blocking=True)

            graph = [given_nds, given_eds]
            ind_fixed_nodes, _ = selectNodesTypes(nd_to_sample, batch_size, nds)
            ind_fixed_nodes = ind_fixed_nodes.to(device, non_blocking=True)
            nd_to_sample = nd_to_sample.to(device, non_blocking=True)
            state = {"masks": real_mks, "fixed_nodes": ind_fixed_nodes}
            z, given_masks_in, given_nds, given_eds = _init_input(graph, state)
            z = z.to(device, non_blocking=True)
            given_masks_in = given_masks_in.to(device, non_blocking=True)
            given_nds = given_nds.to(device, non_blocking=True)
            given_eds = given_eds.to(device, non_blocking=True)

            run_d_step = (cfg.get("d_lr", 0.0) > 0 and batch_idx % cfg.get("d_update_interval", 1) == 0)
            d_loss_value = None
            gp_value = None

            if run_d_step:
                for p in discriminator.parameters():
                    p.requires_grad = True
                optimizer_D.zero_grad()
                with torch.no_grad():
                    gen_mks_d = generator(z, given_m=given_masks_in, given_y=given_nds, given_w=given_eds)
                    if cfg.get("hard_copy_fixed_nodes", False):
                        gen_mks_d = hard_copy_fixed_masks(gen_mks_d, real_mks, ind_fixed_nodes)
                real_validity = discriminator(real_mks, given_y=given_nds, given_w=given_eds, nd_to_sample=nd_to_sample)
                fake_validity_d = discriminator(gen_mks_d.detach(), given_y=given_nds, given_w=given_eds, nd_to_sample=nd_to_sample)
                gp = compute_gradient_penalty(
                    discriminator,
                    real_mks,
                    gen_mks_d.detach(),
                    given_y=given_nds,
                    given_w=given_eds,
                    nd_to_sample=nd_to_sample,
                )
                d_loss = -torch.mean(real_validity) + torch.mean(fake_validity_d) + cfg["lambda_gp"] * gp
                d_loss.backward()
                optimizer_D.step()

                d_loss_value = d_loss.item()
                gp_value = gp.item()
                epoch_d_loss += d_loss_value
                epoch_gp_loss += gp_value
                n_d_steps += 1
            else:
                with torch.no_grad():
                    real_validity = discriminator(real_mks, given_y=given_nds, given_w=given_eds, nd_to_sample=nd_to_sample)

            n_batches += 1

            if batch_idx % cfg["n_critic"] == 0:
                optimizer_G.zero_grad()
                for p in discriminator.parameters():
                    p.requires_grad = False

                z_g = torch.randn(real_mks.shape[0], 128, device=device)
                gen_mks = generator(z_g, given_m=given_masks_in, given_y=given_nds, given_w=given_eds)
                gen_mks_for_d = gen_mks
                if cfg.get("hard_copy_fixed_nodes", False):
                    gen_mks_for_d = hard_copy_fixed_masks(gen_mks, real_mks, ind_fixed_nodes)
                if cfg.get("gan_adv_weight", 0.0) > 0:
                    fake_validity = discriminator(gen_mks_for_d, given_y=given_nds, given_w=given_eds, nd_to_sample=nd_to_sample)
                    g_loss_adv = -torch.mean(fake_validity)
                else:
                    fake_validity = None
                    g_loss_adv = torch.tensor(0.0, device=device)
                if len(ind_fixed_nodes) > 0:
                    recon_loss_fixed, fixed_l1_raw, dice_fixed, fg_ratio_fixed = mask_reconstruction_loss(
                        gen_mks[ind_fixed_nodes],
                        given_masks_in[ind_fixed_nodes, 0],
                        cfg,
                    )
                else:
                    recon_loss_fixed = torch.tensor(0.0, device=device)
                    fixed_l1_raw = torch.tensor(0.0, device=device)
                    dice_fixed = torch.tensor(0.0, device=device)
                    fg_ratio_fixed = torch.tensor(0.0, device=device)
                recon_loss_all, all_l1_raw, dice_all, fg_ratio_all = mask_reconstruction_loss(gen_mks, real_mks, cfg)
                recon_loss_nonfixed, nonfixed_l1_raw, dice_nonfixed, fg_ratio_nonfixed, empty_penalty = nonfixed_reconstruction_loss(
                    gen_mks,
                    real_mks,
                    ind_fixed_nodes,
                    cfg,
                )
                fg_ratio_loss_all = foreground_ratio_loss_tanh(gen_mks_for_d, real_mks)
                g_adv_loss_scaled = cfg["gan_adv_weight"] * g_loss_adv
                g_recon_loss = cfg["gan_recon_weight"] * recon_loss_fixed
                g_nonfixed_recon_loss = cfg.get("gan_nonfixed_recon_weight", 0.0) * recon_loss_nonfixed
                g_empty_penalty_loss = cfg.get("gan_empty_penalty_weight", 0.0) * empty_penalty
                g_fg_ratio_loss = cfg.get("gan_fg_ratio_weight", 0.0) * fg_ratio_loss_all
                g_loss = g_adv_loss_scaled + g_recon_loss + g_nonfixed_recon_loss + g_empty_penalty_loss + g_fg_ratio_loss
                g_loss.backward()
                optimizer_G.step()

                epoch_g_loss += g_loss.item()
                epoch_l1_loss += recon_loss_fixed.item()
                epoch_g_adv_loss += g_loss_adv.item()
                epoch_g_fg_ratio_loss += fg_ratio_loss_all.item()
                epoch_g_nonfixed_recon_loss += recon_loss_nonfixed.item()
                epoch_g_empty_penalty_loss += empty_penalty.item()
                epoch_recon_all_loss += recon_loss_all.item()
                n_g_steps += 1

                # 如果有開 W&B，這裡會持續送出 batch-level 訓練指標。
                train_payload = {
                    "train_batch/d_step": int(run_d_step),
                    "train_batch/g_loss": g_loss.item(),
                    "train_batch/g_adv_loss": g_loss_adv.item(),
                    "train_batch/g_adv_loss_scaled": g_adv_loss_scaled.item(),
                    "train_batch/l1_loss": fixed_l1_raw.item(),
                    "train_batch/recon_loss": recon_loss_fixed.item(),
                    "train_batch/recon_all_loss": recon_loss_all.item(),
                    "train_batch/nonfixed_recon_loss": recon_loss_nonfixed.item(),
                    "train_batch/nonfixed_l1_loss": nonfixed_l1_raw.item(),
                    "train_batch/g_recon_loss": g_recon_loss.item(),
                    "train_batch/g_nonfixed_recon_loss": g_nonfixed_recon_loss.item(),
                    "train_batch/fg_ratio_loss_all": fg_ratio_loss_all.item(),
                    "train_batch/g_fg_ratio_loss": g_fg_ratio_loss.item(),
                    "train_batch/dice_fixed": dice_fixed.item(),
                    "train_batch/dice_all": dice_all.item(),
                    "train_batch/dice_nonfixed": dice_nonfixed.item(),
                    "train_batch/fg_ratio_loss_fixed": fg_ratio_fixed.item(),
                    "train_batch/fg_ratio_loss_nonfixed": fg_ratio_nonfixed.item(),
                    "train_batch/empty_penalty": empty_penalty.item(),
                    "train_batch/g_empty_penalty_loss": g_empty_penalty_loss.item(),
                    "train_batch/real_score": real_validity.mean().item(),
                    "train_batch/fake_score": fake_validity.mean().item() if fake_validity is not None else 0.0,
                    "train_batch/real_foreground_ratio": (real_mks > 0).float().mean().item(),
                    "train_batch/gen_foreground_ratio_raw": (gen_mks > 0).float().mean().item(),
                    "train_batch/gen_foreground_ratio_conditioned": (gen_mks_for_d > 0).float().mean().item(),
                    "train_batch/hard_copy_fixed_nodes": int(cfg.get("hard_copy_fixed_nodes", False)),
                    "train_batch/epoch": epoch + 1,
                }
                if run_d_step:
                    train_payload["train_batch/d_loss"] = d_loss_value
                    train_payload["train_batch/gp"] = gp_value
                log_metrics(train_payload, step=global_step)

                if preview_path is None:
                    preview_path = os.path.join(PREVIEW_DIR, f"epoch_{epoch + 1:03d}.png")
                    render_fixed_preview(generator, fixed_preview_batch, preview_path, hard_copy_fixed=cfg.get("hard_copy_fixed_nodes", False))
                    if wandb_run is not None:
                        wandb_run.log({"train/preview": wandb.Image(preview_path), "trainer/global_step": global_step})

                if run_d_step:
                    pbar.set_postfix(d_loss=d_loss_value, g_loss=g_loss.item(), nonfixed=recon_loss_nonfixed.item(), empty=empty_penalty.item(), gp=gp_value)
                else:
                    pbar.set_postfix(d_step=0, g_loss=g_loss.item(), nonfixed=recon_loss_nonfixed.item(), empty=empty_penalty.item())
            else:
                train_payload = {
                    "train_batch/d_step": int(run_d_step),
                    "train_batch/real_score": real_validity.mean().item(),
                    "train_batch/real_foreground_ratio": (real_mks > 0).float().mean().item(),
                    "train_batch/epoch": epoch + 1,
                }
                if run_d_step:
                    train_payload["train_batch/d_loss"] = d_loss_value
                    train_payload["train_batch/gp"] = gp_value
                log_metrics(train_payload, step=global_step)
                if run_d_step:
                    pbar.set_postfix(d_loss=d_loss_value, gp=gp_value)
                else:
                    pbar.set_postfix(d_step=0)

            global_step += 1

        avg_d = epoch_d_loss / max(n_d_steps, 1)
        avg_g = epoch_g_loss / max(n_g_steps, 1)
        avg_l1 = epoch_l1_loss / max(n_g_steps, 1)
        avg_gp = epoch_gp_loss / max(n_d_steps, 1)
        avg_g_adv = epoch_g_adv_loss / max(n_g_steps, 1)
        avg_g_fg_ratio = epoch_g_fg_ratio_loss / max(n_g_steps, 1)
        avg_g_nonfixed_recon = epoch_g_nonfixed_recon_loss / max(n_g_steps, 1)
        avg_g_empty_penalty = epoch_g_empty_penalty_loss / max(n_g_steps, 1)
        avg_recon_all = epoch_recon_all_loss / max(n_g_steps, 1)
        metrics = {
            "epoch": epoch + 1,
            "avg_d_loss": avg_d,
            "avg_g_loss": avg_g,
            "avg_g_adv_loss": avg_g_adv,
            "avg_l1_loss": avg_l1,
            "avg_recon_loss": avg_l1,
            "avg_recon_all_loss": avg_recon_all,
            "avg_nonfixed_recon_loss": avg_g_nonfixed_recon,
            "avg_empty_penalty": avg_g_empty_penalty,
            "avg_fg_ratio_loss": avg_g_fg_ratio,
            "avg_gp": avg_gp,
            "lr_g": optimizer_G.param_groups[0]["lr"],
            "lr_d": optimizer_D.param_groups[0]["lr"],
        }
        current_best_value = metrics.get(best_metric_name, avg_g)
        is_best = current_best_value < best_g_loss
        if is_best:
            best_g_loss = current_best_value

        paths = save_checkpoint(
            generator,
            discriminator,
            optimizer_G,
            optimizer_D,
            epoch + 1,
            global_step,
            cfg,
            metrics,
            CHECKPOINT_DIR,
            RUN_NAME,
            is_best=is_best,
        )
        log_metrics(
            {
                "train_epoch/avg_d_loss": avg_d,
                "train_epoch/avg_g_loss": avg_g,
                "train_epoch/avg_g_adv_loss": avg_g_adv,
                "train_epoch/avg_l1_loss": avg_l1,
                "train_epoch/avg_recon_loss": avg_l1,
                "train_epoch/avg_recon_all_loss": avg_recon_all,
                "train_epoch/avg_nonfixed_recon_loss": avg_g_nonfixed_recon,
                "train_epoch/avg_empty_penalty": avg_g_empty_penalty,
                "train_epoch/avg_fg_ratio_loss": avg_g_fg_ratio,
                "train_epoch/avg_gp": avg_gp,
                "train_epoch/lr_g": optimizer_G.param_groups[0]["lr"],
                "train_epoch/lr_d": optimizer_D.param_groups[0]["lr"],
                "train_epoch/epoch": epoch + 1,
            },
            step=global_step,
        )
        if is_best:
            upload_wandb_training_artifact(
                wandb_run,
                cfg,
                RUN_NAME,
                generator,
                paths["best"],
                epoch + 1,
                metrics=metrics,
            )
        if wandb_run is not None:
            wandb_run.summary["best_metric"] = best_metric_name
            wandb_run.summary["best_value"] = best_g_loss
            wandb_run.summary["best_checkpoint"] = paths["best"]
            wandb_run.summary["last_checkpoint"] = paths["last"]
            wandb_run.summary["load_pretrained_generator"] = cfg.get("load_pretrained_generator", False)
            wandb_run.summary["pretrained_generator_checkpoint"] = cfg.get("pretrained_generator_checkpoint")

        print(
            f"Epoch {epoch + 1}: avg_d={avg_d:.4f}, avg_g={avg_g:.4f}, "
            f"avg_g_adv={avg_g_adv:.4f}, avg_l1={avg_l1:.4f}, "
            f"avg_nonfixed={avg_g_nonfixed_recon:.4f}, avg_empty={avg_g_empty_penalty:.4f}, "
            f"avg_gp={avg_gp:.4f}, best_{best_metric_name}={best_g_loss:.4f}"
        )

    print("Training complete.")
    print("Best checkpoint:", checkpoint_paths(CHECKPOINT_DIR, RUN_NAME)["best"])
    print("Last checkpoint:", checkpoint_paths(CHECKPOINT_DIR, RUN_NAME)["last"])

    if wandb_run is not None:
        wandb_run.finish()
```

## 13) Simple Diagnostics

```python
def nonfixed_mask_metrics(gen_mks, real_mks, fixed_nodes):
    fixed_mask = torch.zeros(gen_mks.shape[0], dtype=torch.bool, device=gen_mks.device)
    if fixed_nodes is not None and len(fixed_nodes) > 0:
        fixed_mask[fixed_nodes.long()] = True
    nonfixed = torch.arange(gen_mks.shape[0], device=gen_mks.device)[~fixed_mask]
    if len(nonfixed) == 0:
        return {}

    pred = gen_mks[nonfixed] > 0
    target = real_mks[nonfixed] > 0
    intersection = (pred & target).float().sum(dim=(-2, -1))
    union = (pred | target).float().sum(dim=(-2, -1)).clamp_min(1.0)
    pred_area = pred.float().mean(dim=(-2, -1))
    target_area = target.float().mean(dim=(-2, -1))
    return {
        "nonfixed_iou_mean": (intersection / union).mean().item(),
        "nonfixed_pred_area_mean": pred_area.mean().item(),
        "nonfixed_target_area_mean": target_area.mean().item(),
        "nonfixed_empty_rate": (pred_area == 0).float().mean().item(),
        "nonfixed_oversized_rate": (pred_area > target_area * 2.0).float().mean().item(),
    }


metrics = nonfixed_mask_metrics(gen_mks, real_mks, ind_fixed_nodes)
print(metrics)

# Diversity sanity check: same condition, multiple z values.
n_variants = 8
generator_infer.eval()
variants = []
with torch.no_grad():
    for _ in range(n_variants):
        z_var = torch.randn(real_mks.shape[0], 128, device=device)
        sample = generator_infer(z_var, given_m=given_masks_in.to(device), given_y=given_nds.to(device), given_w=given_eds.to(device))
        if cfg.get("hard_copy_fixed_nodes", False):
            sample = hard_copy_fixed_masks(sample, real_mks, ind_fixed_nodes)
        variants.append(sample.detach().cpu())

fixed_set = set(map(int, ind_fixed_nodes.detach().cpu().tolist()))
show_node = next((idx for idx in range(real_mks.shape[0]) if idx not in fixed_set), 0)
fig, axes = plt.subplots(1, n_variants, figsize=(2.0 * n_variants, 2.2))
for i, ax in enumerate(np.array(axes).reshape(-1)):
    ax.imshow(variants[i][show_node].numpy(), cmap="gray", vmin=-1, vmax=1)
    ax.set_title(f"z {i}")
    ax.axis("off")
plt.suptitle(f"Diversity check for node {show_node}")
plt.tight_layout()
plt.show()
```
